Le rapport de classification fournit des mesures de précision, de rappel et de F1-score pour chaque classe, ainsi que d'autres mesures telles que le support. Voici comment interpréter chaque partie du rapport :

Precision (Précision) : La précision est la proportion d'observations positives correctement identifiées parmi toutes les observations identifiées comme positives par le modèle. Par exemple, pour la classe 0, la précision est de 0.78, ce qui signifie que parmi toutes les prédictions faites par le modèle comme étant de la classe 0, 78% d'entre elles sont correctes.

Recall (Rappel) : Le rappel est la proportion d'observations positives correctement identifiées parmi toutes les observations réellement positives dans les données. Par exemple, pour la classe 0, le rappel est de 0.91, ce qui signifie que le modèle a réussi à identifier 91% de tous les véritables exemples de la classe 0.

F1-score : Le F1-score est la moyenne harmonique de la précision et du rappel, ce qui donne une mesure combinée de ces deux métriques. Il est particulièrement utile lorsque les classes sont déséquilibrées. Par exemple, pour la classe 0, le F1-score est de 0.84.

Support : Le support est le nombre d'occurrences de chaque classe dans l'ensemble de test. Par exemple, il y a 80 exemples de la classe 0 et 63 exemples de la classe 1 dans l'ensemble de test.

Accuracy (Exactitude) : C'est la proportion d'échantillons correctement classés parmi tous les échantillons dans l'ensemble de test. Dans ce cas, l'exactitude globale est de 0.80, ce qui correspond à ce que vous avez obtenu précédemment.

Le rapport de classification fournit également des moyennes macro et pondérées pour la précision, le rappel et le F1-score, qui donnent une vue d'ensemble des performances du modèle sur toutes les classes. En général, une macro-avg ou une weighted-avg élevée indique de bonnes performances globales du modèle.
