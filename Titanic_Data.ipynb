Le rapport de classification fournit des mesures de prcision, de rappel et de F1-score pour chaque classe, ainsi que d'autres mesures telles que le support. Voici comment interprter chaque partie du rapport :

Precision (Prcision) : La prcision est la proportion d'observations positives correctement identifies parmi toutes les observations identifies comme positives par le modle. Par exemple, pour la classe 0, la prcision est de 0.78, ce qui signifie que parmi toutes les prdictions faites par le modle comme tant de la classe 0, 78% d'entre elles sont correctes.

Recall (Rappel) : Le rappel est la proportion d'observations positives correctement identifies parmi toutes les observations rellement positives dans les donnes. Par exemple, pour la classe 0, le rappel est de 0.91, ce qui signifie que le modle a russi  identifier 91% de tous les vritables exemples de la classe 0.

F1-score : Le F1-score est la moyenne harmonique de la prcision et du rappel, ce qui donne une mesure combine de ces deux mtriques. Il est particulirement utile lorsque les classes sont dsquilibres. Par exemple, pour la classe 0, le F1-score est de 0.84.

Support : Le support est le nombre d'occurrences de chaque classe dans l'ensemble de test. Par exemple, il y a 80 exemples de la classe 0 et 63 exemples de la classe 1 dans l'ensemble de test.

Accuracy (Exactitude) : C'est la proportion d'chantillons correctement classs parmi tous les chantillons dans l'ensemble de test. Dans ce cas, l'exactitude globale est de 0.80, ce qui correspond  ce que vous avez obtenu prcdemment.

Le rapport de classification fournit galement des moyennes macro et pondres pour la prcision, le rappel et le F1-score, qui donnent une vue d'ensemble des performances du modle sur toutes les classes. En gnral, une macro-avg ou une weighted-avg leve indique de bonnes performances globales du modle.
